```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.figure(figsize=(10, 6))
colors = ["#bac1bd"]*len(avg_artist_streams)# Create a list of colors for the bars

max_index = avg_artist_streams.idxmax() # Find the index of the artist with the highest average streams
highlight_position = avg_artist_streams.index.get_loc(max_index) # Get the position of that artist in the index
colors[highlight_position] = "#bf2517" # Change the color of that artist's bar to a different color

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width = 0.8)  # Top 10 artists
ax = plt.gca() # Get the current axes object
ax.spines['top'].set_visible(False) # Remove the top spine
ax.spines['right'].set_visible(False) # Remove the right spine


plt.xlabel("Artist")# Add a label for the x-axis
plt.ylabel("Average Total Streams (Millions)")# Add a label for the y-axis
plt.title("Average Total Streams by Artist on Spotify (2024)")# Add a title to the chart
plt.xticks(rotation=45, ha="right")# Rotate x-axis labels for better readability

plt.tight_layout() # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
plt.figure(figsize=(12, 8))

special_artist = "Ed Sheeran"
colors = ["#8a6eec" if artist == special_artist else "gray" for artist in avg_artist_streams.index]  # Highlight Ed Sheeran in purple



plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width = 0.9)  # Top 10 artists
ax = plt.gca() # Get the current axes object
ax.spines['top'].set_visible(False) # Remove the top spine
ax.spines['right'].set_visible(False) # Remove the right spine


plt.xlabel("Artist")# Add a label for the x-axis
plt.ylabel("Average Total Streams (Millions)")# Add a label for the y-axis
plt.title("Ed Sheeran Ranks 10th in Total Average Streams(2024)")# Add a title to the chart
plt.xticks(rotation=45, ha="right")# Rotate x-axis labels for better readability

plt.tight_layout() # Adjust layout to prevent clipping of tick-labels
plt.savefig("../plots/EdSheeranBarPlots")
plt.show()
```

```python
catergory ="Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(catergory)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(10,6))# Set the figure size for better readability 10 width and 6 height
plt.pie(
    platform_totals,
    labels=platform_totals.index,
    startangle=110,
    colors = [ "#474749","#33ff66"],
    autopct='%1.1f%%') # Create a pie chart of total streams by platform type
plt.title("Near even split betwenn Free and Premium Spotify (2024)", fontsize = 18, pad=20)# Add a title to the chart
plt.legend(
    labels = platform_totals.index,
    loc="lower right",
    title = "Platform"
)# Add a legend to the pie chart

plt.tight_layout() # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))# Set the figure size for better readability 10 width and 6 height

wedges, texts, autotexts = ax.pie(
    platform_totals,
    autopct='%1.1f%%',
    startangle=110,
    colors = [ "#474749","#33ff66"])# Create a pie chart of total streams by platform type

ax.set_title("Near even split betwenn Free and Premium Spotify (2024)", fontsize = 18, pad=20)# Add a title to the chart

plt.title(
    "Near even split between Free and Premium Users", 
    fontsize = 24, 
    pad=20,
    loc="center") # Add a title to the chart
for autotext in autotexts:
    autotext.set_color('white')  # Set the color of the percentage text to white
    autotext.set_fontsize(20)    # Set the font size of the percentage text
    autotext.set_weight('bold')   # Set the font weight of the percentage text to bold

ax.legend(
    wedges,
    platform_totals.index,
    loc="lower center",
    bbox_to_anchor=(0.5, 0.95),
    ncol=2,
)# Add a legend to the pie chart

plt.tight_layout() # Adjust layout to prevent clipping of tick-labels
plt.savefig("../plots/PieCharts")
plt.show()
```

```python
plt.figure(figsize=(10,6))# Set the figure size for better readability 10 width and 6 height
plt.hist(
    spotify_df["Avg Stream Duration (Min)"],
    bins=20,
    color="#ff7f0e",
    
    edgecolor='white'
)# Create a histogram of average stream duration

plt.xlabel("Average Stream Duration (Minutes)")# Add a label for the x-axis
plt.ylabel("Number of Tracks")# Add a label for the y-axis
plt.title("Distribution of Average Stream Duration", fontsize = 16, pad=20)# Add a title to the chart

ax = plt.gca() # Get the current axes object
for spine in ax.spines.values():
    spine.set_linewidth(0.5)  # Set the spine line width to 0.25
    spine.set_alpha(0.5)  # Set the spine transparency to 0.6
ax.spines['top'].set_visible(False) # Remove the top spine
ax.spines['right'].set_visible(False) # Remove the right spine

plt.savefig("../plots/Histograms")
plt.show()
```

```python
numeric_spotify_df = spotify_df.select_dtypes(include=[np.number])
corr =numeric_spotify_df.corr()

im = plt.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)# Create a heatmap of the correlation matrix
plt.colorbar(im, label ="Pearson Corellation")

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
plt.yticks(range(len(corr.columns)), corr.columns)

plt.savefig("../plots/Heatmaps")
plt.show()
```
