```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
# group by genre and calculate average streams
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values) # Create a bar chart of average streams by genre
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values) # Create a bar chart of average streams by genre

plt.xlabel("Genre")# Add a label for the x-axis
plt.ylabel("Average Total Streams (Millions)")# Add a label for the y-axis
plt.title("Average Total Streams by Genre on Spotify (2024)")# Add a title to the chart

plt.show()
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.figure(figsize=(10, 6))
colors = ["#bac1bd"]*len(avg_artist_streams)# Create a list of colors for the bars

max_index = avg_artist_streams.idxmax() # Find the index of the artist with the highest average streams
highlight_position = avg_artist_streams.index.get_loc(max_index) # Get the position of that artist in the index
colors[highlight_position] = "#bf2517" # Change the color of that artist's bar to a different color

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width = 0.8)  # Top 10 artists
ax = plt.gca() # Get the current axes object
ax.spines['top'].set_visible(False) # Remove the top spine
ax.spines['right'].set_visible(False) # Remove the right spine


plt.xlabel("Artist")# Add a label for the x-axis
plt.ylabel("Average Total Streams (Millions)")# Add a label for the y-axis
plt.title("Average Total Streams by Artist on Spotify (2024)")# Add a title to the chart
plt.xticks(rotation=45, ha="right")# Rotate x-axis labels for better readability

plt.tight_layout() # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
plt.figure(figsize=(12, 8))

special_artist = "Ed Sheeran"
colors = ["#8a6eec" if artist == special_artist else "gray" for artist in avg_artist_streams.index]  # Highlight Ed Sheeran in purple



plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width = 0.9)  # Top 10 artists
ax = plt.gca() # Get the current axes object
ax.spines['top'].set_visible(False) # Remove the top spine
ax.spines['right'].set_visible(False) # Remove the right spine


plt.xlabel("Artist")# Add a label for the x-axis
plt.ylabel("Average Total Streams (Millions)")# Add a label for the y-axis
plt.title("Ed Sheeran Ranks 10th in Total Average Streams(2024)")# Add a title to the chart
plt.xticks(rotation=45, ha="right")# Rotate x-axis labels for better readability

plt.tight_layout() # Adjust layout to prevent clipping of tick-labels
plt.show()
```
